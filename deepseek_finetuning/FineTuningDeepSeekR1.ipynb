```python
# Complete Script for Fine-Tuning DeepSeek Coder, Evaluation, and Download

# 1. Import Libraries
# We begin by importing the necessary libraries and modules.
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, Trainer, TrainingArguments
from datasets import load_dataset, Dataset, DatasetDict # Import Dataset and DatasetDict
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # Import matplotlib for visualization
from huggingface_hub import login # Import login function
# from google.colab import userdata # No longer needed for hardcoded token
import os # Import os for directory creation
# For evaluation metrics
import nltk
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
# Import files module for downloading
from google.colab import files


# Download necessary NLTK data for BLEU and ROUGE
# This will automatically download 'punkt' and 'punkt_tab' if not found, without asking for confirmation.
try:
    # Check if punkt tokenizer is already available
    nltk.data.find('tokenizers/punkt')
    print("'punkt' tokenizer found.")
except LookupError:
    # If not found, download it automatically
    print("'punkt' tokenizer not found, downloading...")
    nltk.download('punkt', quiet=True) # Use quiet=True to avoid interactive prompts
    print("Download complete.")

try:
    # Check if punkt_tab tokenizer is already available (needed by word_tokenize)
    nltk.data.find('tokenizers/punkt_tab')
    print("'punkt_tab' tokenizer found.")
except LookupError:
    # If not found, download it automatically
    print("'punkt_tab' tokenizer not found, downloading...")
    nltk.download('punkt_tab', quiet=True) # Use quiet=True to avoid interactive prompts
    print("Download complete.")


# Set device to GPU if available, otherwise CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Hugging Face Login
# Log in to Hugging Face to access potentially gated models
# WARNING: Hardcoding your token is NOT recommended for security.
# Use Colab secrets or environment variables in production.
try:
    # Hardcoded Hugging Face token - Replace with your actual token if needed
    # User requested hardcoding, but using Colab secrets is the secure way.
    hf_token = "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"
    # Check if the token is not the placeholder before logging in
    # NOTE: Using a placeholder token will likely prevent access to gated models.
    # A basic check if it looks like a real token (starts with hf_ and has more characters)
    if hf_token and hf_token.startswith("hf_") and len(hf_token) > 5:
         login(token=hf_token)
         print("Successfully logged in to Hugging Face using hardcoded token.")
    else:
        print("Using a placeholder or empty token. Login skipped. Model loading might fail for gated models.")

except Exception as e:
    print(f"An error occurred during Hugging Face login: {e}")


# 2. Load the Pre-trained Model and Tokenizer
# Load a smaller DeepSeek model along with its tokenizer from the Hugging Face Hub.
# The larger models like DeepSeek-R1 and DeepSeek-R1-Distill-Llama-8B are likely too large for free Colab GPUs.
# Using the DeepSeek-Coder-1.3B-Base model which is publicly available.
model_name = "deepseek-ai/deepseek-coder-1.3b-base"

print(f"Loading original model: {model_name}")
# Load the original model without quantization
# Using torch_dtype=torch.float16 to potentially reduce memory usage.
try:
    original_tokenizer = AutoTokenizer.from_pretrained(model_name)
    original_model = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype=torch.float16, # Specify dtype
    ).to(device) # Move original model to device
    print("Original model loaded successfully.")
except Exception as e:
    print(f"Error loading original model in float16: {e}")
    print("Attempting to load in default dtype (likely float32). This may require more VRAM.")
    try:
        original_tokenizer = AutoTokenizer.from_pretrained(model_name) # Load tokenizer again if first failed
        original_model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
        print("Original model loaded successfully in default dtype.")
    except Exception as e_default:
        print(f"Error loading original model in default dtype: {e_default}")
        print("Could not load the original model. Please check your GPU memory or try a smaller model.")
        original_model = None # Set to None if loading fails


print(f"Loading model for fine-tuning: {model_name}")
# Load the model for fine-tuning without quantization
# Using torch_dtype=torch.float16 to potentially reduce memory usage.
try:
    tokenizer = AutoTokenizer.from_pretrained(model_name) # Load tokenizer again if first failed
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype=torch.float16, # Specify dtype
    ).to(device) # Move model to device
    print("Model for fine-tuning loaded successfully.")
except Exception as e:
    print(f"Error loading model for fine-tuning in float16: {e}")
    print("Attempting to load in default dtype (likely float32). This may require more VRAM.")
    try:
        tokenizer = AutoTokenizer.from_pretrained(model_name) # Load tokenizer again if first failed
        model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
        print("Model for fine-tuning loaded successfully in default dtype.")
    except Exception as e_default:
        print(f"Error loading model for fine-tuning in default dtype: {e_default}")
        print("Could not load the model for fine-tuning. Please check your GPU memory or try a smaller model.")
        model = None # Set to None if loading fails


# Add a pad token if the tokenizer doesn't have one (necessary for some models and batching)
if 'tokenizer' in locals() and tokenizer is not None and tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})
    if 'model' in locals() and model is not None:
        model.resize_token_embeddings(len(tokenizer))
if 'original_tokenizer' in locals() and original_tokenizer is not None and original_tokenizer.pad_token is None:
     original_tokenizer.add_special_tokens({'pad_token': '[PAD]'})
     if 'original_model' in locals() and original_model is not None:
          original_model.resize_token_embeddings(len(original_tokenizer))


# Check if models were loaded before proceeding
if 'model' not in locals() or model is None or 'original_model' not in locals() or original_model is None:
    print("Model loading failed. Exiting script.")
    # You might want to add a sys.exit(1) here in a real script
else:
    # 3. Data Preparation
    # Load a dataset from Hugging Face for fine-tuning and evaluation
    # Using a subset of the samsum dataset for demonstration
    dataset_name = "samsum"
    print(f"\nLoading dataset: {dataset_name}")

    try:
        # Load the full dataset, explicitly trusting remote code
        full_dataset = load_dataset(dataset_name, trust_remote_code=True)
        print("Dataset loaded successfully.")
        print(full_dataset)

        # Create a smaller subset for faster training and evaluation in Colab
        # Adjust the number of examples as needed
        train_subset_size = 50 # Use 50 examples for training
        eval_subset_size = 10  # Use 10 examples for evaluation

        dataset = DatasetDict({
            "train": full_dataset["train"].select(range(min(train_subset_size, len(full_dataset["train"])))),
            "validation": full_dataset["validation"].select(range(min(eval_subset_size, len(full_dataset["validation"]))))
        })
        print(f"Created smaller dataset subset for training ({len(dataset['train'])} examples) and validation ({len(dataset['validation'])} examples).")
        print(dataset)

    except Exception as e:
        print(f"Error loading dataset {dataset_name}: {e}")
        print("Please check the dataset name, your internet connection, or if 'py7zr' is installed.")
        print("Exiting script as dataset loading failed.")
        # You might want to add a sys.exit(1) here in a real script
        exit() # Exit if dataset loading fails


    def preprocess_function(examples):
        # Preprocess the dataset for text generation (summarization)
        # Input: Dialogue, Target: Summary
        inputs = [f"Summarize the following dialogue:\n{dialogue}\nSummary:" for dialogue in examples['dialogue']]
        targets = examples['summary']

        # Tokenize inputs
        model_inputs = tokenizer(inputs, truncation=True, padding="max_length", max_length=512)

        # Tokenize targets
        # In causal language modeling, labels are the input_ids shifted.
        # We will concatenate input and target and use the combined sequence as input_ids and labels.
        # The loss will be calculated only on the target part by masking the input part later in the Trainer.
        # However, for simplicity here, we'll just tokenize the full text and rely on Trainer's default behavior
        # which calculates loss on all tokens, but the fine-tuning signal will primarily come from the target part.
        # A more precise approach would involve manually creating labels and masking.

        # Concatenate input and target for training
        full_texts = [f"Summarize the following dialogue:\n{dialogue}\nSummary: {summary}" for dialogue, summary in zip(examples['dialogue'], examples['summary'])]
        tokenized_full_texts = tokenizer(full_texts, truncation=True, padding="max_length", max_length=512)

        model_inputs["labels"] = tokenized_full_texts["input_ids"]

        # Add the original summary text to the tokenized dataset for evaluation
        model_inputs["target_text"] = targets

        return model_inputs


    # Apply the preprocessing function to the dataset
    # Remove columns that are not needed by the model to avoid issues with the Trainer
    tokenized_dataset = dataset.map(preprocess_function, batched=True, remove_columns=["dialogue", "summary", "id"])
    print("Dataset tokenized.")
    print(tokenized_dataset)

    # 4. Fine-Tuning Setup (Supervised Fine-Tuning)
    # Configure the training parameters and initialize the Trainer for supervised fine-tuning.

    output_dir = "./deepseek-coder-finetuned-samsum"
    print(f"\nSetting up Trainer with output directory: {output_dir}")

    training_args = TrainingArguments(
        output_dir=output_dir, # Directory to save checkpoints and outputs
        num_train_epochs=3, # Number of training epochs
        per_device_train_batch_size=2, # Reduced batch size for potentially lower VRAM
        per_device_eval_batch_size=2, # Reduced batch size for potentially lower VRAM
        learning_rate=5e-5, # Learning rate for the optimizer
        logging_steps=10, # Log training progress every N steps
        # Corrected argument name
        eval_strategy="steps", # Evaluate the model every N steps
        save_steps=100, # Save a model checkpoint every N steps
        fp16=torch.cuda.is_available(),  # Enable mixed precision (FP16) if GPU is available
        # Corrected argument name
        remove_unused_columns=True, # Set to True to avoid issues with extra columns
        gradient_accumulation_steps=4, # Add gradient accumulation to simulate larger batch size
        eval_steps=10, # Evaluate every 10 steps
        save_total_limit=2, # Limit the number of checkpoints saved
        load_best_model_at_end=True, # Load the best model at the end of training
        metric_for_best_model="eval_loss", # Metric to use for loading the best model
        greater_is_better=False, # For loss, lower is better
    )

    # Initialize the Hugging Face Trainer
    # Need a data collator for padding
    from transformers import DataCollatorForLanguageModeling
    # Using DataCollatorForLanguageModeling with tokenizer.pad_token_id
    data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)


    trainer = Trainer(
        model=model, # The model to train
        args=training_args, # The training arguments
        train_dataset=tokenized_dataset["train"], # The training dataset
        eval_dataset=tokenized_dataset["validation"], # The evaluation dataset
        data_collator=data_collator, # Add data collator
    )

    print("\nStarting supervised fine-tuning...")
    # Run the supervised fine-tuning process:
    try:
        trainer.train()
        print("\nSupervised fine-tuning completed.")

        # 7. Save the Fine-tuned Model
        save_directory = "./fine_tuned_model_output_samsum"
        os.makedirs(save_directory, exist_ok=True)
        print(f"\nSaving fine-tuned model to {save_directory}...")
        trainer.save_model(save_directory)
        print("Fine-tuned model saved.")

        # 8. Download the Fine-tuned Model
        # Corrected indentation for the download block
        print(f"\nAttempting to download the directory: {save_directory}")
        try:
            # Zip the directory for easier download
            zip_filename = "fine_tuned_model.zip"
            # Ensure the zip command is correctly formed and handles potential spaces or special characters
            os.system(f'zip -r "{zip_filename}" "{save_directory}"')


            # Download the zipped file
            files.download(zip_filename)
            print(f"Successfully initiated download of {zip_filename}")

        except Exception as e:
            print(f"An error occurred during the download process: {e}")
            print("Please check the directory path and your Colab environment.")


    except Exception as e:
        print(f"\nError during training: {e}")
        print("Training might fail if the dataset is not properly formatted or if there are memory issues.")
        print("Consider reducing batch size, increasing gradient accumulation, or using a smaller model.")


    # 5. Optional: Reinforcement Learning Fine-tuning (Pseudocode)
    # Note: This section remains pseudocode as in the original document.
    # A full implementation would require specific RL libraries and setup.
    print("\nSkipping Optional Reinforcement Learning Fine-tuning (Pseudocode).")
    print("Refer to the original document for the conceptual RL loop.")


    # 6. Evaluation and Visualization
    # Evaluate on the validation subset
    eval_dataset = tokenized_dataset["validation"]
    # Decode input_ids back to text for evaluation prompts
    # Corrected column name from "input_inputs" to "input_ids"
    eval_prompts = [tokenizer.decode(input_ids, skip_special_tokens=True) for input_ids in eval_dataset["input_ids"]]
    true_summaries = eval_dataset["target_text"]


    print("\nEvaluating models on the validation dataset...")

    # Function to generate output from a model
    def generate_output(model, tokenizer, prompt, max_length=100):
        # Ensure model is in evaluation mode
        model.eval()
        # Move inputs to the correct device
        # Use the tokenizer from the model being evaluated
        inputs = tokenizer(prompt, return_tensors="pt", padding=True, truncation=True, max_length=512).to(model.device)

        # Generate output
        with torch.no_grad(): # Disable gradient calculation for inference
            # Set generation parameters - adjust as needed
            output_ids = model.generate(
                **inputs,
                max_new_tokens=max_length, # Generate up to max_length new tokens
                num_return_sequences=1,
                pad_token_id=tokenizer.pad_token_id,
                eos_token_id=tokenizer.eos_token_id, # Stop generation at EOS token
                do_sample=True, # Use sampling
                top_k=50,
                top_p=0.95,
                temperature=0.7,
            )

        # Decode and return the result
        result = tokenizer.decode(output_ids[0], skip_special_tokens=True)

        # Post-processing: Extract the summary part if the model generates the full prompt+summary
        # This depends on how the model was trained and the prompt format.
        # For the samsum dataset with "Summarize... Summary:", we can split.
        if "Summary:" in result:
             result = result.split("Summary:")[-1].strip()
        # Further cleaning might be needed depending on the model's output
        # e.g., removing unwanted special tokens or trailing text.
        return result

    # Collect outputs from both models
    original_generated_summaries = []
    finetuned_generated_summaries = []

    # Check if original_model was loaded successfully before evaluating
    if 'original_model' in locals() and original_model is not None:
        print("\nGenerating summaries from Original Model:")
        # Use the original tokenizer for the original model
        for i, prompt in enumerate(eval_prompts):
             print(f"Evaluating example {i+1}/{len(eval_prompts)}")
             original_result = generate_output(original_model, original_tokenizer, prompt)
             original_generated_summaries.append(original_result)
             # print(f"Prompt: {prompt}\nOriginal Summary: {original_result}\nTrue Summary: {true_summaries[i]}\n---") # Optional: print for debugging
    else:
        print("\nOriginal model was not loaded successfully, skipping original model evaluation.")


    # Check if finetuned model was loaded successfully before evaluating
    if 'model' in locals() and model is not None:
        print("\nGenerating summaries from Fine-tuned Model:")
        # Use the fine-tuned tokenizer for the fine-tuned model
        model.to(device) # Ensure model is on the correct device
        for i, prompt in enumerate(eval_prompts):
             print(f"Evaluating example {i+1}/{len(eval_prompts)}")
             finetuned_result = generate_output(model, tokenizer, prompt)
             finetuned_generated_summaries.append(finetuned_result)
             # print(f"Prompt: {prompt}\nFine-tuned Summary: {finetuned_result}\nTrue Summary: {true_summaries[i]}\n---") # Optional: print for debugging
    else:
         print("\nFine-tuned model was not loaded successfully, skipping fine-tuned model evaluation.")


    # Calculate Performance Metrics (BLEU and ROUGE)
    print("\nCalculating Performance Metrics (BLEU and ROUGE)...")

    # Initialize scorers
    rouge_scorer_obj = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    smoothie = SmoothingFunction().method4 # Smoothing function for BLEU

    original_rouge_scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}
    original_bleu_scores = []

    finetuned_rouge_scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}
    finetuned_bleu_scores = []


    # Calculate scores for original model outputs
    if original_generated_summaries:
        print("Calculating scores for Original Model...")
        for generated, reference in zip(original_generated_summaries, true_summaries):
            # ROUGE expects a single reference, BLEU can take multiple (as a list of lists)
            scores = rouge_scorer_obj.score(reference, generated)
            original_rouge_scores['rouge1'].append(scores['rouge1'].fmeasure)
            original_rouge_scores['rouge2'].append(scores['rouge2'].fmeasure)
            original_rouge_scores['rougeL'].append(scores['rougeL'].fmeasure)

            # BLEU requires tokenized sentences
            reference_tokens = [nltk.word_tokenize(reference)] # List of reference sentences (here just one)
            generated_tokens = nltk.word_tokenize(generated)
            original_bleu_scores.append(sentence_bleu(reference_tokens, generated_tokens, smoothing_function=smoothie))

        avg_original_rouge = {k: np.mean(v) for k, v in original_rouge_scores.items()}
        avg_original_bleu = np.mean(original_bleu_scores)
        print(f"Average Original Model ROUGE: {avg_original_rouge}")
        print(f"Average Original Model BLEU: {avg_original_bleu:.4f}")


    # Calculate scores for fine-tuned model outputs
    if finetuned_generated_summaries:
        print("Calculating scores for Fine-tuned Model...")
        for generated, reference in zip(finetuned_generated_summaries, true_summaries):
            scores = rouge_scorer_obj.score(reference, generated)
            finetuned_rouge_scores['rouge1'].append(scores['rouge1'].fmeasure)
            finetuned_rouge_scores['rouge2'].append(scores['rouge2'].fmeasure)
            finetuned_rouge_scores['rougeL'].append(scores['rougeL'].fmeasure)

            reference_tokens = [nltk.word_tokenize(reference)]
            generated_tokens = nltk.word_tokenize(generated)
            finetuned_bleu_scores.append(sentence_bleu(reference_tokens, generated_tokens, smoothing_function=smoothie))

        avg_finetuned_rouge = {k: np.mean(v) for k, v in finetuned_rouge_scores.items()}
        avg_finetuned_bleu = np.mean(finetuned_bleu_scores)
        print(f"Average Fine-tuned Model ROUGE: {avg_finetuned_rouge}")
        print(f"Average Fine-tuned Model BLEU: {avg_finetuned_bleu:.4f}")


    # Visualization of ROUGE-L and BLEU
    print("\nGenerating Performance Visualization (ROUGE-L and BLEU)...")

    metrics_labels = ['ROUGE-1', 'ROUGE-2', 'ROUGE-L', 'BLEU']
    original_metric_values = [avg_original_rouge.get('rouge1', 0), avg_original_rouge.get('rouge2', 0), avg_original_rouge.get('rougeL', 0), avg_original_bleu if original_generated_summaries else 0]
    finetuned_metric_values = [avg_finetuned_rouge.get('rouge1', 0), avg_finetuned_rouge.get('rouge2', 0), avg_finetuned_rouge.get('rougeL', 0), avg_finetuned_bleu if finetuned_generated_summaries else 0]

    x = np.arange(len(metrics_labels)) # the label locations
    width = 0.35 # the width of the bars

    fig, ax = plt.subplots(figsize=(10, 6))

    if original_generated_summaries:
        rects1 = ax.bar(x - width/2, original_metric_values, width, label='Original Model')
        # Add value labels on bars
        for rect in rects1:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., height + 0.01,
                    f'{height:.4f}',
                    ha='center', va='bottom')

    if finetuned_generated_summaries:
        rects2 = ax.bar(x + width/2, finetuned_metric_values, width, label='Fine-tuned Model')
        # Add value labels on bars
        for rect in rects2:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., height + 0.01,
                    f'{height:.4f}',
                    ha='center', va='bottom')

    # Add some text for labels, title and axes ticks
    ax.set_ylabel('Score')
    ax.set_title('Model Performance Comparison (BLEU and ROUGE Scores)')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics_labels)
    ax.set_ylim(0, 1) # Scores are typically between 0 and 1
    ax.legend()

    fig.tight_layout()
    plt.show()


    print("\nEvaluation and Visualization complete.")

```
